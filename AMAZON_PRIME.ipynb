```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("amazon_prime_titles.csv")  # Make sure the file name matches yours

# General style
sns.set(style="whitegrid", palette="Set2")
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Movie vs TV Show Count
plt.figure()
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#FF9999','#66B3FF'])
plt.title("Distribution of Movies vs TV Shows on Prime Video")
plt.ylabel("")
plt.tight_layout()
plt.show()

# 2. Top 10 Countries by Number of Titles
plt.figure()
top_countries = df['country'].dropna().str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Content-Producing Countries")
plt.xlabel("Number of Titles")
plt.tight_layout()
plt.show()

# 3. Top 10 Genres
plt.figure()
top_genres = df['listed_in'].dropna().str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres on Prime Video")
plt.xlabel("Number of Titles")
plt.tight_layout()
plt.show()

# 4. Release Year Trend
plt.figure()
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
yearly_counts = df['release_year'].value_counts().sort_index()
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker="o")
plt.title("Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles Released")
plt.tight_layout()
plt.show()

# 5. Content Rating Distribution
plt.figure()
rating_counts = df['rating'].value_counts().head(10)
sns.barplot(x=rating_counts.values, y=rating_counts.index)
plt.title("Top 10 Audience Ratings")
plt.xlabel("Number of Titles")
plt.tight_layout()
plt.show()

# 6. Most Frequent Cast/Directors (Optional - needs heavy parsing)
# Uncomment if you're ready to handle 'cast' or 'director' columns
# top_cast = df['cast'].dropna().str.split(', ').explode().value_counts().head(10)
# sns.barplot(x=top_cast.values, y=top_cast.index)
# plt.title("Most Frequent Cast Members")
# plt.xlabel("Appearances")
# plt.tight_layout()
# plt.show()
```
