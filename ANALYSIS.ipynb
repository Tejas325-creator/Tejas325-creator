```python
# we see that for age if we plot bar graph then it will be not great as its difficult to analyze
# so we use line graph or AGEBINS
```

```python
teco_base_data1['Age'].value_counts().sort_index(ascending=True).plot()
# BY DEFAULT IT PLOTS LINEGRAPH
```

```python
# FOR ANALYSING AGE
Tot=sns.kdeplot(teco_base_data1.Age[teco_base_data1['Exited'==0]],color = 'Red',shades=True)
Tot=sns.kdeplot(teco_base_data1.Age[teco_base_data1['Exited'==1]],color = 'Blue',shades=True)
Tot.legend(['No churn','Churn'],loc = 'upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Age')
Tot.set_title('Age vs Churn')

```

```python
# FOR ANALYSING AGE
# Corrected boolean indexing for filtering
Tot=sns.kdeplot(teco_base_data1.Age[teco_base_data1['Exited']==0],color = 'Red',shade=True)
Tot=sns.kdeplot(teco_base_data1.Age[teco_base_data1['Exited']==1],color = 'Blue',shade=True)
Tot.legend(['No churn','Churn'],loc = 'upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Age')
Tot.set_title('Age vs Churn')
```
